```python

listOf_selected_sectors = list(selected_sectors['industry'].unique())


sectors_data = []

for sector in listOf_selected_sectors:
    tumor_volumes = selected_sectors.loc[selected_sectors['industry'] == sector, 'avg_spread']
    #@ tumor_volumes = combined_data.query("`Drug Regimen` == treatment")['Tumor Volume (mm3)']

    quartiles = tumor_volumes.quantile([.25, .5, .75])
    lowerq = quartiles[0.25]
    upperq = quartiles[0.75]
    iqr = upperq - lowerq

    lower_bound = lowerq - (1.5 * iqr)
    upper_bound = upperq + (1.5 * iqr)
    outliers = tumor_volumes[(tumor_volumes < lower_bound) | (tumor_volumes > upper_bound)]

    sectors_data.append(tumor_volumes)
    print(f"{sector}'s potential outliers: {outliers}")
```

```python
fig, ax = plt.subplots()
ax.boxplot(sectors_data, labels=listOf_selected_sectors)
ax.set_title('Distribution of avg_spread for Each industry')
ax.set_xlabel('industry')
ax.set_ylabel('avg_spread')
plt.xticks(rotation=90)
plt.show()
```
