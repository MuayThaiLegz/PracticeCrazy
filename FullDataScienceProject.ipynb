```python
import pandas as pd
cluster_df = pd.read_excel('Data_4321.xlsx', 'data', index_col = 'A16')
cluster_df.shape
cluster_df.head()
```

```python
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
import matplotlib.pyplot as plt

#load data
data =cluster_df

# Identify categorical columns
categorical_columns = data.select_dtypes(include=['object']).columns

# Preprocess categorical columns using one-hot encoding
ct = ColumnTransformer([('encoder', OneHotEncoder(), categorical_columns)], remainder='passthrough')
X = ct.fit_transform(data)

# Initialize an empty list to store the inertia
inertia = []

# Perform K-means clustering for K values ranging from 1 to 15
for k in range(1, 16):
    kmeans = KMeans(n_clusters=k, random_state=0)
    kmeans.fit(X)
    inertia.append(kmeans.inertia_)

# Plot the inertia values against the number of clusters (K)
plt.plot(range(1, 16), inertia, marker='o')
plt.xlabel('Number of Clusters (K)')
plt.ylabel('Inertia')
plt.title('K-means Clustering')
plt.show()

import pandas as pd
pd.options.mode.chained_assignment = None
```

## Elbow and Silouette

```python
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt

# Load data into a pandas DataFrame
data = cluster_df
# Identify categorical columns
categorical_columns = data.select_dtypes(include=['object']).columns

# Preprocess categorical columns using one-hot encoding
ct = ColumnTransformer([('encoder', OneHotEncoder(), categorical_columns)], remainder='passthrough')
X = ct.fit_transform(data)

# Initialize lists to store the inertia and silhouette scores
inertia = []
silhouette_scores = []


# Perform K-means clustering for K values ranging from 1 to 15
for k in range(2, 16):
    kmeans = KMeans(n_clusters=k, random_state=0)
    kmeans.fit(X)
    inertia.append(kmeans.inertia_)
    silhouette_scores.append(silhouette_score(X, kmeans.labels_))

# Plot the elbow curve
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.plot(range(2, 16), inertia, marker='o')
plt.xlabel('Number of Clusters (K)')
plt.ylabel('Inertia')
plt.title('Elbow Curve')

# Plot the silhouette curve
plt.subplot(1, 2, 2)
plt.plot(range(2, 16), silhouette_scores, marker='o')
plt.xlabel('Number of Clusters (K)')
plt.ylabel('Silhouette Score')
plt.title('Silhouette Curve')

plt.tight_layout()
plt.show()
```

Voting Classifier (VotingClassifier): This is an ensemble meta-model that combines several different models into one classifier. It provides an output prediction based on the majority of voting from all the included classifiers. The code uses 'soft' voting, which means the class labels are predicted based on the predicted probabilities of the classifiers, as opposed to 'hard' voting which uses the mode of the labels.

```python

voting_clf = VotingClassifier(
    estimators=[('sgd', sgd_clf), ('dt', fullClassTree), ('gb', boost)],
    voting='soft')
voting_clf.fit(X_train, y_train)
```

We have three distinct models with comparable AUROC scores in the current scenario: SGD (Stochastic Gradient Descent) with an AUROC of 0.965, Decision Tree with an AUROC of 0.845, and Boosted model with an AUROC of 0.985.

Based on the reported AUROC values, the Boosted model had the greatest AUROC score of 0.985, suggesting the best performance in terms of differentiating between positive and negative occurrences. When compared to the other models, the Boosted model shows high predictive skills and is likely to deliver more accurate predictions.

The SGD model had an AUROC score of 0.965, which is likewise fairly high and shows good classification accuracy. While the AUROC score of 0.965 is significantly lower than that of the Boosted model, it indicates that the SGD model is similarly good at differentiating between positive and negative occurrences.

The Decision Tree model had the lowest AUROC score, which was 0.845. Despite having the lowest score of the three models, it is crucial to remember that Decision Trees might still be effective in some situations. * Final Evaluation

The final testing phase in machine learning involves using a completely separate set of data, called the test set, to evaluate the performance of your final model. This phase is essential because it gives an unbiased estimate of how well your model will perform on unseen data in the future.

In the final testing phase, the test set, which the model has never seen during the training or validation phase, is used. This ensures that the model's performance metrics, like accuracy, precision, recall, F1 score reflect how the model would perform on new, unseen data.
